# 5. Perform meaningful visualizations using the dataset. 

# 5.1 Violin Plot

```python
plt.figure(figsize=(20,20))
sns.violinplot(x='Classification',y='PL',data=df)
```

# 5.2 Swarm Plot

```python
plt.figure(figsize=(10,10))
sns.swarmplot(x='Classification',y='PL',data=df,hue='Classification')
```

# 5.3 Strip Plot

```python
plt.figure(figsize=(10,10))
sns.stripplot(x='Classification',y='SL',data=df,hue='Classification')
```

# Strip plot & Swarn plot together to see the difference

```python
plt.figure(figsize=(10,10))
sns.stripplot(x='Classification',y='SL',data=df)
sns.swarmplot(x='Classification',y='SL',data=df)
```

# 5.4 Dis Plot

```python
plt.figure(figsize=(15,15))
plt.style.use('Solarize_Light2')
sns.displot(data=df,x ='SW',hue='Classification',height=8,multiple='stack')
```
